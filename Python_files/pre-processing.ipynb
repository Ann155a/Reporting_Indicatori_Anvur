## Trasformazione Wide

L'obiettivo delle rappresentazioni che verranno prodotte di seguito a questa fase di pre-processing, saranno le *classe di laurea*, pertanto questa è l'unità delle analisi. Essendo tale, dobbiamo trasformare il dataset, in modo da avere questo dato nelle righe, e le descrizioni (mappate prima) nelle colonne come feature. \
Inoltre, nel dataset abbiamo ulteriori informazioni sulle feature per ogni classe di laurea, anche a livello del *Sud* e di tutta *l'Italia*. \
Per questo motivo, ci saranno 3 righe per ogni classe di laurea che riportano dati per ciascun feature in base al tipo di indicatore. 

```python
df.columns.to_list()
```

```python
print(f'Questo dataset ha {df_2024.shape[0]} righe e {df_2024.shape[1]} colonne.')
```

### Tabella Pivot
Il dataset verrà trasformato da un formato *long* a uno *wide* per correggere anche le problematiche legate alla maggiore dimensionalità dello spazio delle feature.

```python
df_long = df.melt(
    id_vars=['CLASSE DI LAUREA', 'NOME_CORSO', 'DIPARTIMENTO', 'DESCRIZIONE'],
    value_vars=['INDICATORE_CDS', 'INDICATORE_ITA', 'INDICATORE_SUD'],
    var_name='INDICATORE_TYPE',
    value_name='VALUE'
)
```

```python
df_long
```

```python
# pivot DESCRIZIONE in colonne
df_pivot = df_long.pivot_table(
    index=['CLASSE DI LAUREA', 'INDICATORE_TYPE', 'NOME_CORSO', 'DIPARTIMENTO'],
    columns='DESCRIZIONE',
    values='VALUE'
).reset_index()
```

```python
df_pivot
```

```python
print(f'Questo dataset ha {df_pivot.shape[0]} righe e {df_pivot.shape[1]} colonne.')
```

```python
df_pivot.columns.to_list()
```

## Features selezionate

```python
colonne_utili = [
    # Info corso
    'CLASSE DI LAUREA', 'NOME_CORSO', 'DIPARTIMENTO', 'INDICATORE_TYPE',

    # Studenti
    'Iscritti_totali',

    # Laureati
    'Laureati_in_tempo', 'Laureati_totali', 

    # Occupazione
   'Percentuale_Occupazione_3anni_LM',

    # Prosecuzione studi
    'Percentuale_Proseguono_II_anno', 

    # Docenti / qualità
    'Ore_docenza_indeterminato', 'Qualita_ricerca_QRDLM', 'Percentuale_Soddisfazione_studenti', 
]
```

```python
df_pivot = df_pivot[colonne_utili]
```

```python
df_imputed = df_pivot.copy() # no imputation
```

```python
# Colonne testuali da capitalizzare
cols_to_capitalize = [
    'CLASSE DI LAUREA',
 'INDICATORE_TYPE',
 'NOME_CORSO',
 'DIPARTIMENTO'
]

for col in cols_to_capitalize:
    df_imputed[col] = df_imputed[col].str.strip().str.capitalize()

# Sistema INDICATORE_TYPE
df_imputed["INDICATORE_TYPE"] = (
    df_imputed["INDICATORE_TYPE"]
    .str.upper()
    .str.replace("INDICATORE_", "", regex=False)
    .str.strip()
)
```
